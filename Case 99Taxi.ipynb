# Case 99Taxi

### Objetivo:
- Encontrar a melhor alocação de analistas de call center, dividindo-os em 2 canais de atendimento: Driver e Pax.

### Dados:
- Horário de entrada de 369 analistas de call center
- Histórico de tickets (atendimentos) agrupados em horário cheio, durante 7 dias, para os canais Driver e Pax.

```python
import scipy.stats as ss
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import functools
import re
```

Os seguintes arquivos `.csv` foram extraídos do dataset original, tendo sido separados em `Driver.csv`, que contém os `tickets recebidos via Driver`, e em `Pax.csv`, que contém os `tickets recebidos via Pax`.

```python
pd_driver = pd.read_csv('Driver.csv', delimiter=';')
pd_pax = pd.read_csv('Pax.csv', delimiter=';')
```

```python
pd_driver.columns
```

```python
pd_pax.columns
```

Criando duas grandes listas, contendo respectivamente, todas as chamadas recebidas pelo canal Driver e Pax.

```python
list_driver_calls = list(pd_driver['CHAMADAS QUE ENTRARAM NA FILA'])
```

```python
list_pax_calls = list(pd_pax['CHAMADAS QUE ENTRARAM NA FILA'])
```

# Preparando os dados para gerar a distribuição de poisson para chamadas via Driver

Separando o volume de chamadas (tickets) por dia, em listas distintas:

```python
lista_24 = list_driver_calls[0:24]
lista_25 = list_driver_calls[24:48]
lista_26 = list_driver_calls[48:72]
lista_27 = list_driver_calls[72:96]
lista_28 = list_driver_calls[96:120]
lista_29 = list_driver_calls[120:144]
lista_30 = list_driver_calls[144:168]
```

Encontrando a média de volume de atendimentos, para as chamadas de cada um dos dias:

```python
mean_24 = functools.reduce(lambda a,b: a+b, lista_24)/24
mean_25 = functools.reduce(lambda a,b: a+b, lista_25)/24
mean_26 = functools.reduce(lambda a,b: a+b, lista_26)/24
mean_27 = functools.reduce(lambda a,b: a+b, lista_27)/24
mean_28 = functools.reduce(lambda a,b: a+b, lista_28)/24
mean_29 = functools.reduce(lambda a,b: a+b, lista_29)/24
mean_30 = functools.reduce(lambda a,b: a+b, lista_30)/24
list_mean = [mean_24, mean_25, mean_26, mean_27, mean_28, mean_29, mean_30]
```

Esta lista gerada contém a média do número de ligações recebidas via Driver, para cada um dos 7 dias

```python
list_mean
```

Verificando a proporção da média diária em relação a média total (considerando os 7 dias)

```python
add = 0
for l in list_mean:
    add += l
    
for l in list_mean:
    print(((l/add)*100))
```

## Criando a distribuição de Poisson para todas as médias para cada um dos 7 dias de chamadas 99 Taxi via Driver

```python
total_calls = []
list_total_calls = []

for mean in list_mean:
    for i in range(1000):
        random_calls = ss.poisson.rvs(mu=mean, size=24)
        total_calls.append(random_calls)
    list_total_calls.append(total_calls)
    total_calls = []

#Esta variável possui uma lista contendo todas as gerações randomicas de numeros ao redor do mean
list_total_calls

```

# Gerando proporções de analistas necessários por dia para atender via Pax

```python
lista_proporcoes_pax = []
soma_pax = dia_24_pax_max + dia_25_pax_max + dia_26_pax_max + dia_27_pax_max + dia_28_pax_max + dia_29_pax_max + dia_30_pax_max
lista_dias_pax = [dia_24_pax_max, dia_25_pax_max, dia_26_pax_max, dia_27_pax_max, dia_28_pax_max ,dia_29_pax_max, dia_30_pax_max]

for dia in lista_dias_pax:
    lista_proporcoes_pax.append(dia/soma_pax)
lista_proporcoes_pax
```

# Calculando a proporção de atendimentos necessários por dia

As variáveis `lista_dias` (contém a lista de médias + 2 std de atendimentos para Driver) e `lista_dias_pax` (contém a lista de médiaz + 2 std de atendimentos para Pax).

As variáveis `soma` (contém a soma de todas as médias + 2 std de atendimentos para Driver e `soma_pax` (contém a soma de todas as médias + 2 std de atendimentos para Pax).

```python
lista_prop = []

for i in range(len(lista_dias)):
    lista_prop.append((lista_dias[i] + lista_dias_pax[i]) / (soma + soma_pax))
```

# Calculando o numero de atendentes que serão necessários por dia

Para este cálculo, considerei o `número de analistas` de call center fornecidos, que são `369`, e `multipliquei por 5`, pois eles trabalham em turnos `5 x 2` (trabalham 5 dias, folgam 2). 

Desta forma, distribuí o número de analistas gerados, pelos `7 dias`, `considerando a proporção` obtida pela média relativa (de cada um dos dias) de `atendimentos` considerando o volume total de atendimentos nos 7 dias.

```python
# Cada atendente de call center trabalha 5 dias na semana e folga 2, portanto
# multiplicaremos o número de analistas pelo numero de dias trabalhados na semana
num_atendentes = 369*5  

# vamos gerar uma lista de distribuição de atendentes para cada dia (24,25,26,27,28,29,30)
lista_distr_atend = []

for l in lista_prop:
    lista_distr_atend.append(int(round(l * num_atendentes,0)))

# vamos imprimir o numero ideal de atendentes por dia 24 --> 30
print('A alocação de analistas por dia (considerando Driver e Pax juntos): ', lista_distr_atend)
```

# Agora vou calcular a proporção total de chamadas entre Driver e Pax

```python
lista_prop_driver = []
lista_prop_pax = []

for i in range(len(lista_dias)):
    lista_prop_driver.append((lista_dias[i]) / (lista_dias[i] + lista_dias_pax[i]))
    lista_prop_pax.append((lista_dias_pax[i]) / (lista_dias[i] + lista_dias_pax[i]))
```

# Aplicando as proporções de Driver e Pax para a lista de distribuição de analistas de call center

```python
alocacao_ideal_driver = []
alocacao_ideal_pax = []

for i in range(len(lista_distr_atend)):
    alocacao_ideal_driver.append(int(round(lista_prop_driver[i] * lista_distr_atend[i])))
    alocacao_ideal_pax.append(int(round(lista_prop_pax[i] * lista_distr_atend[i])))
```
